```python
#앞은 or로 시작하고 뒤는 ge로 끝나는 걸 찾고 싶다?

urls = '''http://www.naver.com/
https://mail.naver.com/
ftp://ftp.naver.com/'''
```

```python
#전방탐색 없이 앞에 프로토콜만 불러오기위해서는?
url_list = urls.split()

for url in url_list:
    protocol = re.match(r'.+\://', url).group()
    protocol = protocol[:-3]
    print(protocol)
```

```python
for url in url_list:
    protocol = re.match(r'.+(?=://)',url).group()
    print(protocol)
```

```python
amounts = '''
ABC01: ₩23.45
ABC01: $23.45
HGG42: $5.31
HGG42: €5.31
CFMX1: $899.00
CFMX1: ¥899.00
'''
```

```python
#문자 나누기 접근
amount_list = amounts.strip().split("\n")

for amount in amount_list:
    money = re.search(r'(?<=: ).+',amount).group()
    money = money[1:]
    print(money)

```

```python
# findall 후방탐색 접근

for amount in amount_list:
    if amount:
        print(re.findall(r'(?<=: .).+',amount)[0])
```

```python
text = '''
<p>
<span>네이버가 뉴스 서비스에 인공지능(AI)을 도입해 페이지 뷰(PV)를 늘리고 이용자를 끌어 모으고 있다.</span>
<span>네이버는 5일 오전 서울 강남구 그랜드 인터컨티넨털 호텔에서 AI 콜로키움 2019를 열고 이 같은 AI 성과와 전략을 소개했다.</span>
<span>이날 기조연설에서 김광현 네이버 서치앤클로바 리더는 "AI 뉴스 추천 시스템인 에어스(AiRS)를 도입하면서 뉴스 소비량이 확대되고 있다" 고 말했다.</span>
</p>
'''
```

```python
span_list = re.findall(r'<span>.+</span>',text)
new_list = []

for span in span_list:
    result = re.findall(r'(?<=<span>).+(?=</span>)',span)  #replace로 없애도 됨, 직접 슬라이싱 해도 됨!
    new_list.append(result[0])
    
new_list
```

```python
phones ='''
park 010-1234-1234
kim 02-3450-3459
lee 000000
'''
```

```python
#전화번호와 이름을 가져오고 싶다
#re.search(r'[a-zA-Z]+\s\d{3}-\d{3,4}-\d{4}',phone)

for phone in phones.split('n'):   #\s는 띄어쓰기나 엔터같은 눈에 보이지 않는 것들을 포함
    if phone:
        num = re.search(r'([a-zA-Z]+)\s(\d{3}-\d{3,4}-\d{4})',phone).group()
        print("이름 : ", num.split()[0])
        print("전화번호 : ", num.split()[1])
        
        
```

```python
print(re.findall(r'([a-zA-Z]+)\s(\d{3}-\d{3,4}-\d{4})',phone))  # 그룹지은 것은 튜플로 반환
```
