```python
import re
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
```

### 실습03 네이버 뉴스 크롤링

```python
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'
}
```

```python

page = 1
yesterday = (datetime.today() - timedelta(1)).strftime("%Y%m%d")

while True:
    url = f"https://news.naver.com/main/list.nhn?mode=LPOD&mid=sec&oid=003&date={yesterday}&page={page}"
    response = requests.get(url,headers=headers)
    soup = BeautifulSoup(response.text,"html.parser")

    li_list1 = soup.select('.type06_headline li')
    li_list2 = soup.select('.type06 li')
    li_list = li_list1 + li_list2
    page += 1
    
    for li in li_list:
        body_url = li.select_one('a')['href']
        body_response = requests.get(body_url, headers=headers)
        body_soup = BeautifulSoup(body_response.text, "html.parser")
    
        if (body_soup.find(id='articeBody') == None) == True and (body_soup.find(id='newsEndContents') == None) == True:
            article = body_soup.find(id='articleBodyContents') #연예, 스포츠 아닌 사회기사
            try:
                print(li.select_one('img')['alt']) #기사 제목 출력
                print("")
                print(re.sub(r"[\n\t]","",article.text.strip())) #본문 내용 출력
                print("")
                print("")
            except:
                print(li.select_one('a').text.strip())

        
        elif (body_soup.find(id='articeBody') == None) == True:
            article = body_soup.find(id='newsEndContents') #스포츠 기사
            print(li.select_one('img')['alt'])
            print("")
            print(re.sub(r"[\n\t]","",article.text.strip()))
            print("")
            print("")
    
        else:
            article = body_soup.find(id='articeBody')
            print(li.select_one('img')['alt']) #연예 기사
            print("")
            print(re.sub(r"[\n\t]","",article.text.strip()))
            print("")
            print("")

    if len(li_list) < 20:
        break
    
    
```
