```python
#실습01 1번  number 간편하게 가져오는 법

url = "https://finance.naver.com/sise/sise_quant.nhn"
response = requests.get(url)
soup = BeautifulSoup(response.text,"html.parser")

list1 = soup.select(".tltle")
number1 = soup.select(".number")
result = []

for i in range(len(list1)):
    result.append(f"{list1[i].text} {number1[i*10].text}")

for j, v in enumerate(result):
    print(j+1, v)
```

```python
#실습01 1번 행 접근으로 다시!!

url = "https://finance.naver.com/sise/sise_quant.nhn"
response = requests.get(url)
soup = BeautifulSoup(response.text,"html.parser")

tr_list = soup.select(".type_2 > tr")  # tr은 종목 수  # td는 종목 이름, 현재가, 전일비 포함
result = []
result_plus = []

for tr in tr_list:
    if len(tr.select("td")) != 12:
        pass
    elif len(tr.select("td")) == 12:
        index = tr.select("td")[1].text
        price = tr.select("td")[2].text
        yester = tr.select("td")[3].text.strip()
        result.append(index + " " + price)
        result_plus.append(index + " " + price + " " + yester)
    else:
        break

for i, v in enumerate(result):
    print(i+1,v)
```

```python
#실습01 2번

tr_list = soup.select(".type_2 > tr")  # tr은 종목 수  # td는 종목 이름, 현재가, 전일비 포함
up_result = []

for tr in tr_list:
    if len(tr.select("td")) != 12:
        pass
    elif len(tr.select("td")) == 12:
        try:
            if tr.select("td")[3].select_one('img')['alt'] == '상승':
                up_index = tr.select("td")[1].text
                up_price = tr.select("td")[2].text
                up_yester = tr.select("td")[3].text.strip()
                up_result.append(up_index + " " + up_price + " " + up_yester)
        except:
            pass


for i, v in enumerate(up_result):
    result_index = result_plus.index(v)
    print(result_index+1,result_plus[result_index])
```
